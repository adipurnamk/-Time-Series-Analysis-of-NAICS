# Data Wrangling


```python
import pandas as pd
import glob
```

```python
csv_list = glob.glob('./*.csv')
csv_list 
```

```python
# Make function to check column
def column_check(index):

    columns = pd.read_csv(csv_list[index]).columns
    for i in csv_list[index:index+12]:
        df = pd.read_csv(i)
        if columns.all() == df.columns.all():
            print(i, 'columns is identical')
        else:
            print(i, 'columns is NOT identical')
```

### Column Check

```python
column_check(1)
```

```python
# Concat 2NAICS dataframes

_2NAICS_list = []

for csv in csv_list[:5]:
    df = pd.read_csv(csv)
    df = df.groupby(['SYEAR', 'NAICS']).sum()
    _2NAICS_list.append(df)
    
_2NAICS_df = pd.concat(_2NAICS_list)
_2NAICS_df.drop('SMTH', axis=1, inplace=True)
_2NAICS_df.sort_values('SYEAR', inplace=True)
_2NAICS_df
```
